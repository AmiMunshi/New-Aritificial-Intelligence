Minimax search for the given tree


Minimax.png

```python
#Define the leaf nodes and total levels
terminal= [[3,12,8], [2,4,6], [14,5,2]]
totalLevels=2
```

```python
#Function that will return minimum of the nodes
def minimizer(nodeList):
  return min(nodeList)
```

```python
#Function that will return maximum of the nodes
def maximizer(nodeList):
  return max(nodeList)
```

```python
#Function to either call minimizer or maximizer at each level
def minimax(level, nodes):
#If level is odd then call minimizer else maximizer
  if level%2!=0:
    new=[]
    for i in nodes:
      if type(i)== list:
#If the nodes has lists within lists then send individual elements of the list to the function
        new.append(minimizer(i))
      else:
#If nodes does not have any more lists within, then send the nodes itself to the fucntion
        new.append(minimizer(nodes))
        break
  else:
    new=[]
    for i in nodes:
      if type(i)== list:
        new.append(maximizer(i))
      else:
        new.append(maximizer(nodes))
        break
    
  return new
```

```python
#initialize the nodes and levels
nodes= terminal
level=totalLevels-1
while level >-1:
  nodes= minimax(level,nodes)
  level-=1
print(nodes) #will print the value at the root node
```
